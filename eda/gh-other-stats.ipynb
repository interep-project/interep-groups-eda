# GitHub Other Stats EDA
## Data Source
GitHub API (REST & GraphQL)

## Data Loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_json('../data/gh-user-stats.json', orient='index')
df.drop('login', axis='columns', inplace=True)
```

## Univariate Data Exploration

```python
df.describe()
```

## Visualization

```python
df.columns
```

### Histograms

```python
fig = plt.figure()
for i, stat in enumerate(df.columns):
    ax = fig.add_subplot(2, 3, i + 1)
    df[stat].hist(bins=3, ax=ax)
    ax.set_title(f"{stat} distribution")
    
fig.tight_layout()
plt.show()
```

### Boxplots

```python
%matplotlib inline
df.boxplot(figsize=[10,6])
```

## Quantiles

```python
df.quantile([0.6, 0.9, 0.95, 0.99, 0.999])
```
